```python
#checking the event SQL symphony concert
query_18 = '''
SELECT person_id, count(*), event_name
FROM facebook_event_checkin
GROUP BY person_id
having count(*) = 3 AND event_name = "SQL Symphony Concert" AND date like "%201712%";
'''
pd.read_sql_query(query_18, con)
```

```python
#Two suspects... Let’s get the names of these people for reference.
query_19 = '''
SELECT DISTINCT gfn_check_in.membership_id, name
FROM get_fit_now_member gfn_member
JOIN get_fit_now_check_in gfn_check_in
ON gfn_member.id = gfn_check_in.membership_id
WHERE gfn_check_in.membership_id IN("48Z7A", "48Z55");
'''
pd.read_sql_query(query_19,con)
```

```python
#Find which of the two suspects has a license id that contains "H42W".
query_20 = '''
SELECT dl.id, p.name, plate_number FROM drivers_license dl
JOIN person p
ON dl.id = p.license_id
JOIN get_fit_now_member gfn_member
ON p.id = gfn_member.person_id
JOIN get_fit_now_check_in gfn_check_in
ON gfn_member.id = gfn_check_in.membership_id
WHERE gfn_check_in.membership_id IN("48Z7A", "48Z55");
'''
pd.read_sql_query(query_20,con)
```

```python
#Find who orchestrated the murder.
query_21 = '''
SELECT * FROM drivers_license dl
JOIN person p
ON dl.id = p.license_id
JOIN facebook_event_checkin fb_event
ON p.id = fb_event.person_id
WHERE hair_color = "red"AND
gender = "female" AND
car_make = "Tesla" AND
car_model = "Model S";
'''
pd.read_sql_query(query_21,con)
```
